```python
# Question 3

# Answer 3 -
from scipy.stats import norm
import numpy as np
import pandas as pd

def pdf_calculation(x,mean,std):
    z_score=(x-mean)/std
    return z_score

lst=np.random.randn(10,1)

print("The Random Data Set is: ",lst,"\r\n")

df=pd.DataFrame(lst, columns=["Random Column"])
mean = float(df.mean())
std=float(df.std())

pdf_calc=norm.pdf(lst,loc=mean,scale=std)
print("The PDF value of the Data Set is: ",pdf_calc,"\r\n")

first_val=float(lst[0][0])
first_val

final_z_score=pdf_calculation(x=first_val,mean=mean,std=std)
print("Z-score value of the first value of the DataSet is: ", final_z_score,"\r\n")

p_value=norm.sf(abs(final_z_score))*2
print("P-Value of the first value of the Dataset is",p_value)
```

```python
# Question 4

# Answer 4 - The properties of the binomial distribution are: 
# There are two possible outcomes: true or false, success or failure, yes or no. 
# There is 'n' number of independent trials or a fixed number of n times repeated trials.
# The probability of success or failure remains the same for each trial.

# Examples -
# 1. Taking a survey of positive and negative reviews from the public for any specific product or place.
# 2. By using the YES/ NO survey, we can check whether the number of persons views the particular channel.
# 3. The number of votes collected by a candidate in an election is counted based on 0 or 1 probability.
```

```python
# Question 5

# Answer 5 -

import scipy.stats as stat
import pylab 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_data(lst):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(lst,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(lst,dist='norm',plot=pylab)
    plt.show()

lst=np.random.binomial(1,0.4,1000)

plot_data(lst)
```

```python
# Question 6

# Answer 6 -
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
  
# No of Data points
N = 50
  
# initializing random values
data = np.random.randn(N)
  
# getting data of the histogram
count, bins_count = np.histogram(data, bins=10)
  
# finding the PDF of the histogram using count values
pdf = count / sum(count)
  
# using numpy np.cumsum to calculate the CDF
# We can also find using the PDF values by looping and adding
cdf = np.cumsum(pdf)
  
# plotting PDF and CDF
plt.plot(bins_count[1:], pdf, color="red", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
plt.legend()
```

```python
# Question 7

# Answer 7 -

# Binomial distribution is the one in which the number of outcomes are only two, that is success or failure.
# Example of binomial distribution: Coin toss.
# Poisson distribution is the one in which the number of possible outcomes has no limits.
# It describes the probability of certain events occurring in some time interval.
# Example of Poisson distribution: Number of insurance claims per day on an insurance company.
```

```python
# Question 8

# Answer 8 -
import numpy as np
import pandas as pd

s = np.random.poisson(5, 1000)
df=pd.DataFrame(s)
mean=float(df.mean())
var=float(df.var())

print("The Poisson Distribution DataSet is: ",df)

print("Mean of the Data Set is: ",mean)
print("Variance of the Data Set is: ",var)
```

```python
# Question 9

# Answer 9 -

# In Poisson distribution the mean and variance are same and is equal to which is (Lambda *  t) 
# where, Lambda= expected number of events to occur at every time interval
# t = time interval
# An important feature of the Poisson distribution is that the variance increases as the mean increases.

# The mean of the binomial distribution is always equal to p, and the variance is always equal to pq/N.
# where, N = the size of the sample, p = the probability of a successful outcome, q = 1 - p
```
