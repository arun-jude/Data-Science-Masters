```python
# Question 2
# python program to print Python Exception Hierarchy

import inspect

def treeClass(cls, ind = 0):
    
      # print name of the class
    print ('-' * ind, cls.__name__)
      
    # iterating through subclasses
    for i in cls.__subclasses__():
        treeClass(i, ind + 3)


print("Hierarchy for Python Exception Class is : ")
  
# inspect.getmro() Return a tuple 
# of class  cls’s base classes.
  
# building a tree hierarchy 
inspect.getclasstree(inspect.getmro(Exception))

treeClass(Exception)
```

```python
# Question 4
# Lookup Error

# LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.
# We can use LookupError exception class to handle both IndexError and KeyError exception classes.

import inspect

def treeClass(cls, ind = 0):
    
      # print name of the class
    print ('-' * ind, cls.__name__)
      
    # iterating through subclasses
    for i in cls.__subclasses__():
        treeClass(i, ind + 3)


print("Hierarchy for LookupError Class is : ")
  
# inspect.getmro() Return a tuple 
# of class  cls’s base classes.
  
# building a tree hierarchy 
inspect.getclasstree(inspect.getmro(LookupError))

treeClass(LookupError)

print("\r\n")

# 1. example of KeyError
dict1={"name":"arun","number":456789,"e-mail_id":"xyz@gmail.com"}

try:
    print(dict1["Name"])
except KeyError as e:
    print(f"Example 1: KeyError: Dictionary Key Not Found: {e}", {e.__class__})

print("\r\n")
#2.example of IndexError

x = [1, 2, 3, 4]
try:
    print(x[10])
except IndexError as e:
    print(f"Example 2: IndexError: Index Not Found: {e}", {e.__class__})
```
