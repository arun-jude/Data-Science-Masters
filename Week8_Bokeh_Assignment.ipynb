```python
# Question 1
# Answer 1 -  Bokeh plot using python code

import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure , output_file, show
from bokeh.sampledata.iris import flowers

output_file('flower.html')

p = figure(title = 'Flower Graph')
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.scatter(flowers['petal_length'] , flowers['petal_width'])
show(p)
```

```python
# Question 2
# Answer 2 -
# A glyph is a vectorized graphical shape or marker that is used to represent your data.

# A Few Categories of Glyphs:

# 1. Marker: Shapes like circles, diamonds, squares and triangles. Effective for scatter and bubble charts.
# 2. Line: Single, step and multi-line shapes. For building line charts.
# 3. Bar/Rectangle: Traditional or stacked bar (hbar) and column (vbar) charts as well as waterfall or gantt charts.

# example

import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure , output_file, show
from bokeh.sampledata.iris import flowers

output_file('flower_glyph_example.html')

p = figure(title = 'Flower Glyph Graph')
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.circle(flowers['petal_length'] , flowers['petal_width'],color='red')
show(p)
```

```python
# Question 3
# Answer 3 - Axis, title and legend details customization in bokeh plot

x = [3,4,5,6,7]
y = [4,5,6,7,4]
output_file("line.html")
p = figure(title = "line graph")
p.scatter(x,y , size = 20,fill_color = 'red',legend_label = 'legend')
show(p)
```

```python
# Question 4
# Answer 4 -
# The Bokeh server is a component of Bokeh that allows you to build interactive web applications that are connected to Python code running on a server.

#The primary purpose of the Bokeh server is to synchronize data between the underlying Python environment and the BokehJS library running in the browser.

# Bokeh server makes it easy to create interactive web applications that connect front-end UI events to running Python code.

# Bokeh creates high-level Python models, such as plots, ranges, axes, and glyphs, and then converts these objects to JSON to pass them to its client library, BokehJS.

# However, keeping these models in sync between the Python environment and the browser would provide further powerful capabilities:

# 1. respond to UI and tool events in the browser with computations or queries using the full power of Python

# 2. automatically push server-side updates to the UI elements such as widgets or plots in the browser

# 3. use periodic, timeout, and asynchronous callbacks to drive streaming updates
```
