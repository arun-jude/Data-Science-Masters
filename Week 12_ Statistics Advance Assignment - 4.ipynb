```python
# Question 1

# Answer 1 -

# The t-test and the z-test are both statistical tests used for hypothesis testing and making inferences about population parameters 
# based on sample data. However, they differ in their assumptions and when they are appropriate to use.

# Z-Test:

#The z-test is used when the population standard deviation (sigma) is known. It is often employed for large sample sizes (n is 
# typically greater than or equal to 30) because, with a large sample, the sample mean tends to follow a normal distribution 
# due to the Central Limit Theorem.

# Example Scenario for a Z-Test:
# Suppose you are testing whether a new medication has a significant effect on lowering blood pressure. 
# You have access to a large sample of blood pressure measurements for patients before and after taking the medication. 
# Since you have a large sample size and the population standard deviation of blood pressure is known, you can use a z-test to determine 
# whether there is a statistically significant difference in mean blood pressure before and after taking the medication.

# T-Test:

# The t-test is used when the population standard deviation (sigma) is unknown or needs to be estimated from the sample data. 
# It is typically preferred for small sample sizes (n is usually less than 30) because the t-distribution accounts 
# for the increased uncertainty associated with estimating the population standard deviation.

# Example Scenario for a T-Test:
# Consider a scenario where you are comparing the test scores of two different teaching methods on a small group of students.
# You want to determine if there is a significant difference in the mean test scores between the two methods.
# Since you have a small sample size and the population standard deviation is unknown,
# you would use a t-test to compare the means of the two groups.
```

```python
# Question 2

# Answer 2

# One-tailed and two-tailed tests are terms used to describe the directionality of a statistical hypothesis test. 
# They refer to whether the test is focused on a specific direction of effect or difference or if it is concerned with deviations 
# in either direction. The choice between these types of tests depends on the research question and the specific hypothesis being tested.

# One-Tailed Test:

# In a one-tailed test, the alternative hypothesis (H1) is focused on a specific direction of effect or difference. 
# The critical region of the test is located entirely in one tail of the probability distribution.
# One-tailed tests are often used when there is a clear expectation or prior knowledge about the direction of the effect.

# For example, suppose a company introduces a new manufacturing process to reduce the weight of a product. 
# The research question might be whether the new process results in a decrease in weight. 
# In this case, a one-tailed test would be appropriate because the focus is on a specific direction (decrease in weight).

# Two-Tailed Test:

# In a two-tailed test, the alternative hypothesis (H1) is concerned with deviations in either direction from the null hypothesis. 
# The critical region of the test is split into two tails of the probability distribution.
# Two-tailed tests are used when there is no specific expectation about the direction of the effect or when both directions are of interest.

# Continuing with the manufacturing process example, if the research question is whether the new process has any effect
# on the weight (increase or decrease), a two-tailed test would be appropriate because both directions of change are being considered.

# Key Differences:

# - One-Tailed Test:
#  - Focuses on a specific direction of effect or difference.
#  - Critical region is located in one tail of the distribution.
#  - Used when there is a clear expectation or hypothesis about the direction of the effect.
#  - Provides greater statistical power to detect an effect in a specific direction.

#- Two-Tailed Test:
#  - Considers deviations in either direction from the null hypothesis.
#  - Critical region is split into two tails of the distribution.
#  - Used when there is no specific expectation about the direction of the effect or when both directions are of interest.
#  - May be more conservative and requires larger sample sizes to achieve the same level of significance.

# In summary, the choice between a one-tailed and a two-tailed test depends on the research question and the hypotheses being tested.
# One-tailed tests are used when there is a specific expectation about the direction of the effect,
# while two-tailed tests are used when deviations in both directions from the null hypothesis are of interest.
```

```python
# Question 3

# Answer 3 -

# Type 1 and Type 2 errors are concepts associated with hypothesis testing that involve making incorrect decisions 
# about the null hypothesis (H0) and the alternative hypothesis (H1). 
# These errors represent the trade-off between the risk of mistakenly rejecting a true null hypothesis and the risk of failing to 
# reject a false null hypothesis.

# 1. Type 1 Error (False Positive):
#   - A Type 1 error occurs when the null hypothesis (H0) is wrongly rejected when it is actually true.
# In other words, the test indicates a significant result when there is no true effect or difference.
#   - The probability of committing a Type 1 error is denoted by (alpha), the chosen significance level.
#   - Type 1 errors are also known as "false positives."

#Example Scenario for a Type 1 Error:
# Suppose a medical researcher is testing a new drug for effectiveness in reducing blood pressure. 
# The null hypothesis (H0) states that the drug has no effect.
# The researcher conducts a hypothesis test and finds a significant result, leading to the rejection of (H0). 
# However, in reality, the drug has no effect on blood pressure, and the significant result is a Type 1 error.

# 2. Type 2 Error (False Negative):
#   - A Type 2 error occurs when the null hypothesis (H0) is wrongly not rejected when it is actually false. 
# In other words, the test fails to detect a true effect or difference when there is actually a true effect or difference.
#   - The probability of committing a Type 2 error is denoted by (beta).
#   - Type 2 errors are also known as "false negatives."

# Example Scenario for a Type 2 Error:
# Consider a scenario where a quality control inspector is testing a batch of products to determine if they meet a certain 
# weight specification. The null hypothesis (H0) states that the products meet the weight specification.
# The inspector performs a hypothesis test but fails to find a significant result, leading to the failure to reject (H0). 
# However, in reality, the products do not meet the weight specification, and the failure to detect this is a Type 2 error.
```

```python
# Question 5

# Answer 5 - 

# A confidence interval is a range of values that is constructed around a sample statistic (such as the mean or proportion) 
# to estimate the true population parameter with a certain level of confidence. It provides a measure of the precision and 
# uncertainty of the estimate. The confidence interval indicates the range within which the true population parameter is likely to lie.

# Confidence intervals are typically expressed as a range with an associated confidence level, which represents the probability
# that the interval contains the true parameter. For example, a 95% confidence interval means that if the same sampling procedure
# is repeated many times, the interval will contain the true parameter about 95% of the time.

# Formula to calculate a confidence interval for the population mean (mu) using the formula for a z-confidence interval, 
# assuming a normal distribution:

# # Confidence Interval = x + (z (alpha/2) * (sigma/ (n)**1/2))

# Confidence Interval = x - (z (alpha/2) * (sigma/ (n)**1/2))

# Where, x=sample mean; 
# z (alpha/2) = critical value from z-table for alpha/2 (significance value)
# sigma= population standard deviation
# n= sample size

# Example:
# Suppose you want to estimate the average height of a population. You take a random sample of 100 individuals and measure their heights. 
# The sample mean (x) is 170 cm and the population standard deviation (sigma) is 5 cm. Confidence level is 95% 
# alpha = 0.05

# Confidence Interval = x + (z (alpha/2) * (sigma/ (n)**1/2)) = 170 + (1.96 *(5/(100)**1/2)) = 170 + 0.98 = 170.98

# Confidence Interval = x - (z (alpha/2) * (sigma/ (n)**1/2)) = 170 - (1.96 *(5/(100)**1/2)) = 170 - 0.98 = 169.02

# Cofidence Interval = (169.02, 170.98)

# The interpretation of this 95% confidence interval is that we are 95% confident that the 
# true average height of the population falls between 169.02 cm and 170.98 cm. 
```

```python
# Question 8

# Answer 8 -

# The margin of error in a confidence interval is a measure of the uncertainty or variability associated with the estimate of a 
# population parameter (such as a mean or proportion) based on a sample. It represents the range within which the true population 
# parameter is likely to lie, with a certain level of confidence. In other words, it quantifies how much the sample estimate might 
# differ from the actual population parameter.

# The margin of error is usually expressed as a positive value and is related to the width of the confidence interval.
# A narrower confidence interval indicates a smaller margin of error and greater precision in estimating the population parameter.

# Margin of Error (E) = (z (alpha/2) * (sigma/ (n)**1/2)) 
# where,
# z (alpha/2) = critical value from z-table for alpha/2 (significance value)
# sigma= population standard deviation
# n= sample size

# Effect of Sample Size on Margin of Error:

# 1. As the sample size (n) increases, the margin of error (E) decreases.
# 2. Larger sample sizes result in more precise estimates and narrower confidence intervals, which means the range within which the 
#   true population parameter is likely to lie becomes smaller.
# 3. This reduction in margin of error is a result of the Central Limit Theorem, which states that the distribution of the
#   sample mean becomes approximately normal as the sample size increases, allowing for more accurate estimation of the 
#  population parameter.

# Example Scenario:

# Suppose you are conducting a survey to estimate the average income of households in a city. 
# You collect data from two different sample sizes: n1=100 ; n2=500; confidence interval = 95%; alpha = 0.05
# assume population standard deviation is known to be 10,000

# for n1=100; margin of error (E1) = 1.96 * (10000/(100)**1/2))= 1960
# for n2=500; margin of error (E2) = 1.96 * (10000/(500)**1/2))= 876

# In this example, a larger sample size results in a smaller margin of error (E2) compared to a smaller sample size (E1).
# This demonstrates how increasing the sample size leads to a more precise estimate of the 
# population parameter and a narrower confidence interval.
```
