```python
# Question 8

# Answer 8 -

import pandas as pd

dict1={"A":[3,8,6,2,9],
       "B":[5,2,9,3,1],
       "C":[1,7,4,5,2]
      }

def my_query(df1):
    lst=[]
    final_lst=[]
    index_lst=[]

    i=0
    for index, rows in df1.iterrows():
        lst=list(rows)
        if lst[0]>5 and lst[1]<10:
            final_lst.append(lst)
            index_lst.append(i)
        i+=1
        
    df=pd.DataFrame(final_lst,index=index_lst)
  
    return df


df1=pd.DataFrame(dict1)

# printing the condition of column A>5 and column B<10 using pondas query function

df2=df1.query('A>5 and B<10')

print(df2.head())

# printing the condition of column A>5 and column B<10 using a created query function

df3=my_query(df1)

df3.rename(columns = {0:'A',1:'B',2:'C'}, inplace = True)
df3.head()
```

```python
# Question 9
# Answer 9 -

import pandas as pd
import math

dict1={'Values':[10,20,3,4,5,6,7,8,9]}

def my_func(df1):
    summation=0

    lst=[]
    std_lst=[]
    for index, rows in df1.iterrows():
        
        summation+=int(rows)
        lst.append(float(rows))
        
    # Mean calculation methodology
    mean=(summation/len(df1))
    print("Mean of the DataSet is: ", mean,"\r\n")
    
    # Median calculation methodology
    if len(lst)%2==0:
        mid_index=len(lst)//2
        mid_index_before=mid_index-1
        median=(lst[mid_index]+lst[mid_index_before])/2
        print("Median of the DataSet is: ",median,"\r\n")
    else:
        mid_index=len(lst)//2
        median=lst[mid_index]
        print("Median of the DataSet is: ",median,"\r\n")
    # Standard Deviation calculation methodology
    for i in lst:
        std_lst.append((i-mean)**2)
    
 
    
    std_summation=0
    for i in std_lst:
        std_summation+=i

    std_dev=math.sqrt(std_summation/(len(df1)-1))
    print("Standard Deviation of the DataSet is: ",std_dev,"\r\n")
   
        
df=pd.DataFrame(dict1)



df['Values'].describe()
my_func(df)

#pandas function to calculate the mean, median & standard deviation of 'Values' Column

print("Printing Mean, Median and Standard Deviation using Pandas function")

print(df['Values'].describe())

df
```
