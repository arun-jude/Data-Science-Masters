```python
# Question 1

# Answer 1 -
# Estimation in statistics refers to the process of using sample data to make inferences about population parameters.
# Estimation involves calculating estimates, which are values that serve as educated guesses or approximations of unknown 
# population parameters based on the information obtained from a sample. There are two main types of estimation: point estimate and interval estimate.

# 1. Point Estimate:
# A point estimate is a single value that is used to estimate an unknown population parameter. 
# It provides a "best guess" of the parameter based on the available sample data. 
# Point estimates are simple and easy to calculate, but they do not provide information about the precision or variability of the estimate.

# Example:
# Suppose you want to estimate the average height of all students in a school. 
# We randomly select a sample of 50 students and find that their average height is 165 cm. 
# The point estimate of the population mean height is 165 cm.

# 2. Interval Estimate (Confidence Interval):
# An interval estimate, also known as a confidence interval, provides a range of values within which the true population parameter 
# is likely to lie, along with a level of confidence. It accounts for the uncertainty associated with estimation by providing a range 
# rather than a single point.

#Example:
# Using the same example of estimating the average height of students, you calculate a 95% confidence interval for the population mean height 
# as 160 cm to 170 cm. This means that you are 95% confident that the true population mean height falls within this interval.

# Key Differences:

# - Point Estimate: Provides a single value that serves as the estimate of the population parameter.
# - Interval Estimate (Confidence Interval): Provides a range of values (interval) along with a specified level of confidence 
# about the parameter's location.
```

```python
# Question 5 -

# Answer 5 -

import numpy as np

import scipy.stats as stats

group1 = np.array([14, 15, 15, 16, 13, 8, 14, 17, 16, 14, 19, 20, 21, 15, 15, 16, 16, 13, 14, 12])
group2 = np.array([15, 16, 13, 14, 14, 8, 13, 19, 19, 14, 17, 21, 18, 16, 13, 15, 13, 16, 15, 13])

# Null hypothesis is population means are same

print(np.var(group1), np.var(group2))

stats.ttest_ind(a=group1, b=group2, equal_var=True)

#since calculated p-value is 0.956 which is greater than alpha of 0.05. We fail to reject the null hypothesis and since the p-value is strongly
# closer to 1. Hence, there is a high probability that the population means are equal
```

```python
# Question 6 -

# Answer 6 -
# In hypothesis testing, the null hypothesis (H0) and the alternative hypothesis ((H1) are two competing statements or hypotheses 
# about a population parameter. These hypotheses are formulated based on the research question or problem being investigated. 
# The null hypothesis typically represents the status quo or the absence of an effect, while the alternative hypothesis
# represents what the researcher is trying to prove or find evidence for.

# Here are the definitions of the null and alternative hypotheses along with some examples:

# 1. Null Hypothesis (H0):
# The null hypothesis is a statement that there is no effect, no difference, or no relationship in the population. 
# It often includes words like "equal," "no change," "no effect," or "no difference."

# Example 1: 
# (H0): The average weight of a certain product is 500 grams.

# Example 2:
# (H0): There is no significant difference in exam scores between two teaching methods.

# 2. Alternative Hypothesis ((H1) or :
# The alternative hypothesis is a statement that contradicts the null hypothesis and asserts a specific effect, 
# difference, or relationship in the population. It is what the researcher is attempting to find evidence for.

# Example 1: 
# (H1): The average weight of a certain product is not 500 grams.

# Example 2:
# (H1): There is a significant difference in exam scores between two teaching methods.

# In many cases, the alternative hypothesis can be one-tailed or two-tailed:

#- A one-tailed alternative hypothesis specifies the direction of the effect, such as "greater than" or "less than." For example:
#  (H1): The new treatment reduces pain more effectively than the standard treatment.

#- A two-tailed alternative hypothesis does not specify a particular direction of the effect, but it asserts that there is a difference or effect. 
#  For example:
#  (H1): The mean response time is different under the two conditions.

# In hypothesis testing, the goal is to collect sample data and analyze it to determine whether the evidence supports the rejection 
# of the null hypothesis in favor of the alternative hypothesis. The decision is based on statistical analysis and a chosen significance 
# level (alpha), which determines the threshold for accepting or rejecting the null hypothesis.
```

```python
# Question 7

# Answer 7 -

# Hypothesis testing involves a systematic process of evaluating and making decisions about a population parameter based on sample data. 
# The following are the general steps involved in hypothesis testing:

#1. Formulate Hypotheses:
#   - State the null hypothesis (H0): This represents the default assumption or status quo about the population parameter. 
#    It often asserts that there is no effect, no difference, or no relationship.
#   - State the alternative hypothesis (H1): This represents the hypothesis you want to test or the claim you're trying to support. 
#    It contradicts the null hypothesis and asserts a specific effect, difference, or relationship.

#2. Collect Data:
#   - Obtain a representative sample from the population of interest. The size of the sample should be determined based on statistical considerations.

#3. Choose a Significance Level (alpha):
#   - Select a predetermined significance level, often denoted as (alpha). Common choices for alpha include 0.05 or 0.01. This value determines the 
# threshold for making decisions about the null hypothesis.

#4. Select a Statistical Test:
#   - Choose an appropriate statistical test based on the nature of the data, research question, and hypotheses. Different tests are used for 
#    different types of data and scenarios (e.g., t-test, ANOVA, chi-squared test, etc.).

# 5. Calculate Test Statistic:
#   - Calculate the appropriate test statistic based on the sample data and the chosen statistical test. The test statistic quantifies how 
#   far the sample result is from what is expected under the null hypothesis.

# 6. Determine the Critical Region:
#   - Determine the critical region(s) of the test statistic based on the chosen significance level (alpha) and the distribution of the 
#   test statistic under the null hypothesis.

# 7. Calculate P-Value:
#   - Calculate the p-value associated with the test statistic. The p-value represents the probability of observing a test statistic 
#   as extreme as the one calculated, assuming the null hypothesis is true.

# 8. Make a Decision:
#   - Compare the calculated p-value with the significance level (alpha).
#   - If p-value <= (alpha), reject the null hypothesis in favor of the alternative hypothesis.
#   - If p-value > (alpha), do not reject the null hypothesis.

# 9. Draw a Conclusion:
#   - Interpret the results of the hypothesis test in the context of the problem. 
#    State whether there is sufficient evidence to support the alternative hypothesis or if the data do not provide enough evidence 
#   to reject the null hypothesis.

# 10. Report Results:
#   - Clearly communicate the findings of the hypothesis test, including the conclusion, p-value, test statistic, and any relevant effect sizes 
    # or confidence intervals.
```

```python
# Question 8

# Answer 8 -
# The p-value, short for "probability value," is a crucial concept in hypothesis testing. 
# It represents the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from the sample data, 
# assuming that the null hypothesis is true. In other words, the p-value quantifies the strength of evidence against the null hypothesis.

#Key points about the p-value and its significance in hypothesis testing:

# 1. Interpreting the p-value:
#   - A low p-value (typically less than the chosen significance level, (alpha)suggests that the observed data are unlikely under the assumption 
#   of the null hypothesis. This leads to a rejection of the null hypothesis in favor of the alternative hypothesis.
#   - A high p-value indicates that the observed data are relatively likely under the assumption of the null hypothesis. This suggests that there 
#   is insufficient evidence to reject the null hypothesis.

#2. Comparing to the Significance Level (alpha):
#   - The p-value is compared to the chosen significance level ((alpha)to make a decision about the null hypothesis.
#   - If p-value <= (alpha), it is considered statistically significant, and the null hypothesis is rejected.
#   - If p-value > (alpha), it is not statistically significant, and the null hypothesis is not rejected.

# 3. Smaller p-value, Stronger Evidence:
#   - A smaller p-value indicates stronger evidence against the null hypothesis. It suggests that the observed data are unlikely to have 
#   occurred by random chance alone.

#4. Larger p-value, Weaker Evidence:
#   - A larger p-value suggests weaker evidence against the null hypothesis. It implies that the observed data are relatively more consistent 
#   with what would be expected under the null hypothesis.

# 5. Continuous Scale:
#   - The p-value is a continuous value between 0 and 1. A p-value of 0 indicates that the observed data are impossible under the null hypothesis, 
#   while a p-value of 1 indicates that the observed data are fully consistent with the null hypothesis.

# 6. No Absolute Proof:
#   - It's important to note that the p-value does not provide a definitive answer or proof. It only quantifies the evidence in the data 
#   against the null hypothesis.
#    - A non-significant p-value does not prove the null hypothesis is true; it simply suggests that the data do not provide strong evidence 
#   against it.

# 7. Context Matters:
#    - The significance of a p-value depends on the specific research question, the chosen significance level, and the practical implications
#    of the results.

# In summary, the p-value is a critical tool in hypothesis testing that helps researchers make informed decisions about the null hypothesis 
# based on the evidence provided by the sample data. It provides a quantitative measure of the strength of evidence against the null hypothesis 
# and guides the interpretation of research findings.
```

```python
# Question 11

# Answer 11 -
# The Student's t-distribution, often simply referred to as the t-distribution, is a probability distribution that arises in statistical 
# inference when dealing with small sample sizes or when the population standard deviation is unknown. It is a fundamental concept in 
# hypothesis testing and confidence interval estimation, particularly when working with small samples.

# When to use the t-distribution:

# 1. Small Sample Size: The t-distribution is especially useful when working with small sample sizes (typically n < 30) where the sample data 
# may not fully represent the population.

# 2. Unknown Population Standard Deviation: If the population standard deviation (sigma) is unknown and needs to be estimated from the sample data, 
# the t-distribution should be used.

# 3. Hypothesis Testing: When performing hypothesis tests involving the mean of a population based on a small sample, the t-distribution is used 
# to calculate critical values and p-values.

# 4. Confidence Intervals: For estimating the population mean within a certain range with a specified level of confidence, the t-distribution 
# is used to construct confidence intervals.

# 5. Comparing Means: When comparing means from two small samples, such as in the two-sample t-test, the t-distribution is employed to determine 
# whether the observed difference is statistically significant.

# 6. Regression Analysis: In linear regression analysis, when making inferences about regression coefficients or prediction intervals, the 
# t-distribution can be used for small sample sizes.

# In summary, the Student's t-distribution is a valuable tool in statistical analysis, especially when working with small sample sizes and 
# situations where the population standard deviation is unknown. It allows for accurate hypothesis testing, confidence interval estimation, 
# and making informed inferences about population parameters based on limited sample data.
```

```python
# Question 12 

# Answer 12 -
# The t-statistic, also known as the t-score, is a test statistic that measures the difference between a sample statistic and a 
# population parameter in units of the standard error. It is a key component of hypothesis testing and is used to assess whether the 
# difference between the sample data and the null hypothesis is statistically significant.

# The formula for calculating the t-statistic: -

# t= (x- mu)/((s)/ (n)**1/2)
# Where:

# x is the sample mean,

# Mu is the hypothesized population mean under the null hypothesis,

# s is the sample standard deviation,

# n is the sample size.
```
