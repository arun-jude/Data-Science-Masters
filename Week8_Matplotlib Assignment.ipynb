```python
# Question 2

# Answer 2 - 

# Scatter Plot: -
# A scatter plot uses dots to represent values for two different numeric variables.
# The position of each dot on the horizontal and vertical axis indicates values for an individual data point.
# Scatter plots are used to observe relationships between variables.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.figure(figsize=(10,5))
plt.scatter(x,y,c='r',alpha=0.9)
plt.xlabel("This is the X Label")
plt.ylabel("This is the Y Label")
plt.title("Scatter Plot to identify relationship between X & Y Variables")
plt.grid()
```

```python
# Question 3

# Answer 3 -
# Subplot function in matplotlib-
# With the subplot() function you can draw multiple plots in one figure.

import numpy as np
import matplotlib.pyplot as plt

# Line 1
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])


plt.subplot(2, 2, 1)
plt.plot(x,y)

# Line 2
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2, 2, 2)
plt.plot(x,y)

# Line 3

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2,2,3)
plt.plot(x,y)

# Line 4

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,4)
plt.plot(x,y)

plt.show()
```

```python
# Question 5

# Answer 5 -
# Box Plot - Box plots provide a quick visual summary of the variability of values in a dataset. 
# They show the median, upper and lower quartiles, minimum and maximum values, and any outliers in the dataset. Outliers can reveal mistakes or unusual occurrences in data.

# Uses of Box PLot: -

# 1.The outliers and their values
# 2. Symmetry of Data
# 3. Tight grouping of data
# 4. Data skewness – if, in which direction and how

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

box3=[box1,box2]

plt.figure(figsize=(6,3))
plt.boxplot(box1)
plt.show()

plt.figure(figsize=(6,3))
plt.boxplot(box2)
plt.show()

plt.figure(figsize=(8,3))
plt.boxplot(box3)
plt.show()
```
