```python
# Question 1

# Answer 1 - 5 plots that can be plotted using seaborn library: -

# 1. Line Plot - It Is The Visualization Of A Single Function. This Plot Helps Us To See The Relationship Between X-Axis, Y-Axis

# 2. Scatter Plot - Scatter Plot Is The Same As A Line Plot. In a line plot the points are joined by a line.
#    In a scatter plot the Points Are Shown Individually With A Dot, Circle, Or Any Other Shape.

# 3. Histogram Plot - It shows The Underlying Frequency Distribution Of A Set Of Continuous Data

# 4. Bar Plot - It is A Way Of Comparing A Set Of Categorical Data.

# 5. Pair Plot - It creates A Grid Of Axis Such That Each Numeric Variable In Data Will Create A Plot Between Each Other
```

```python
# Question 2

# Answer 2 -
import seaborn as sns
import matplotlib.pyplot as plt
fmri=sns.load_dataset('fmri')

sns.lineplot(x=fmri.timepoint,y=fmri.signal,size=fmri.event, style=fmri.region)
plt.show()
```

```python
# Question 3

# Answer 3 -
import seaborn as sns
import matplotlib.pyplot as plt

titanic=sns.load_dataset('titanic')

sns.boxplot(x=titanic.pclass,y=titanic.age)
plt.show()
sns.boxplot(x=titanic.pclass, y=titanic.fare)
plt.show()
```

```python
# Question 4

# Answer 4 -
import seaborn as sns
import matplotlib.pyplot as plt

diamonds=sns.load_dataset('diamonds')
sns.displot(x=diamonds.price,hue=diamonds.cut)
plt.show()
```

```python
# Question 5

# Answer 5 -
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset('iris')
sns.pairplot(data=iris,hue='species')
plt.show()
```

```python
# Question 6

# Answer 6 -
import seaborn as sns
import matplotlib.pyplot as plt

flights=sns.load_dataset('flights')

sns.heatmap(data=(flights.year,flights.passengers))
plt.show()
```
