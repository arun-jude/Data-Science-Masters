```python
# Question 1

# Answer 1 -
# Probability Mass Function (PMF):
# The PMF is a function that gives the probability of each possible outcome of a discrete random variable. 
# It maps each value of the random variable to its corresponding probability.

# Example:
# Consider rolling a fair six-sided die. The PMF of the random variable "X" (the outcome of the roll) is:

# PMF(X = 1) = 1/6
# PMF(X = 2) = 1/6
# PMF(X = 3) = 1/6
# PMF(X = 4) = 1/6
# PMF(X = 5) = 1/6
# PMF(X = 6) = 1/6
# Here, the PMF tells us the probabilities of rolling each number on the die.

# Probability Density Function (PDF):
# The PDF is a function that describes the likelihood of a continuous random variable falling within a certain range of values. Unlike the PMF, which assigns probabilities to specific values, the PDF provides probabilities for ranges of values.

# Example:
# Consider the height of adult individuals. The height is a continuous random variable.
```

```python
# Question 2

# Answer 2 -

# The Cumulative Density Function (CDF) is a fundamental concept in probability and statistics. 
# It describes the probability that a random variable takes on a value less than or equal to a given value. 
# In other words, the CDF gives you the cumulative probability distribution of a random variable.

# Mathematically, for a random variable X, the CDF is denoted as F(x) and is defined as:

# F(x)=P(X<=x)

# In the case of a discrete random variable, the CDF is the sum of the probabilities of all values less than or equal to x. 
# For a continuous random variable, the CDF is often expressed as an integral of the probability density function (PDF) up to x.

# Example:
# Let's consider rolling a fair six-sided die again. The CDF for the outcome of the roll (X) can be calculated as follows:

# For X = 1: F(1) = P(X <= 1) = 1/6
# For X = 2: F(2) = P(X <= 2) = 2/6
# For X = 3: F(3) = P(X <= 3) = 3/6
# For X = 4: F(4) = P(X <= 4) = 4/6
# For X = 5: F(5) = P(X <= 5) = 5/6
# For X = 6: F(6) = P(X <= 6) = 6/6 = 1

# Why CDF is used:
# 1. Probability Calculation: The CDF provides a way to calculate the probability of a random variable falling within a certain range 
# or below a specific value.
# 2. **Quantiles and Percentiles**: The CDF helps in finding quantiles (values that divide the distribution into specified proportions) 
# and percentiles (values that correspond to specific percent levels).
# 3. Distribution Characteristics: The shape and behavior of the CDF reveal important information about the distribution of the random variable, 
# such as skewness, central tendency, and spread.
# 4. Statistical Analysis: The CDF is used in hypothesis testing, confidence interval estimation, and other statistical procedures.
# 5. Comparison: CDFs can be used to compare different distributions and assess their relative likelihoods.
```

```python
# Question 5 

# Answer 5 -

#  Bernoulli distribution is a discrete probability distribution that models a random experiment with two possible outcomes: 
# success (usually denoted as 1) and failure (usually denoted as 0), where the probability of success is denoted by p and the 
# probability of failure is q=(1−p).

# Difference between Bernoulli Distribution and Binomial Distribution:

# 1. Number of Trials:
#   - Bernoulli Distribution: Represents a single trial or experiment with two possible outcomes (success or failure).
#   - Binomial Distribution: Represents the number of successes in a fixed number of independent and identical trials.

# 2. Number of Outcomes:
#   - Bernoulli Distribution: Only has two possible outcomes: 0 (failure) or 1 (success).
#   - Binomial Distribution: Can have multiple possible outcomes, ranging from 0 to the total number of trials.

# 3. Parameters:
#   - Bernoulli Distribution: Has a single parameter (p), the probability of success.
#  - Binomial Distribution: Has two parameters: (n), the number of trials, and (p), the probability of success in each trial.
    
# 4. Probability Mass Function (PMF):
#   - Bernoulli Distribution: The PMF specifies the probability of a single success (1) or failure (0).
#   - Binomial Distribution: The PMF gives the probability of observing a specific number of successes in \(n\) trials.    
```

```python
# Question 7

# Answer 7 -

# The uniform distribution is a probability distribution that describes a scenario where all possible outcomes are equally likely 
# over a specified range. In other words, in a uniform distribution, each value within the range has the same probability of occurring. 
# It's often visualized as a flat, constant probability density function.

# Example of Uniform Distribution:
# Consider rolling a fair six-sided die. The outcomes (1, 2, 3, 4, 5, and 6) are equally likely, and each has a probability of (1/6) of occurring.

# In this case, the uniform distribution is described by:

# Lower bound (a): 1 (minimum value on the die)
# Upper bound (b): 6 (maximum value on the die)
# The probability density function for the uniform distribution on this interval is: 
# f(x)= (1/6), for 1<=x<=6
# This means that each value between 1 and 6 (inclusive) has an equal probability of (1/6) of being rolled
```

```python
# Question 8

# Answer 8 -

# Z-score, also known as a standard score, is a statistical measure that quantifies the number of standard deviations 
# a data point is from the mean of a distribution. It is calculated using the formula:
# z = (x - Mu)/(Sigma/ (n)1/2)
# where, x= sample mean; Mu= Population mean; Sigma=Population standard deviation; n= sample size

# Importance of the z-score:

# 1. Standardization: Z-scores standardize data, allowing for meaningful comparisons between different distributions.
# 2. Normal Distribution: Z-scores are crucial for working with the standard normal distribution (z-distribution),
# which simplifies calculations and statistical analysis.
# 3. Outlier Detection: Z-scores help identify outliers by quantifying how far a data point deviates from the mean.
# 4. Probability Calculation: Z-scores are used to calculate probabilities and determine the percentile rank of a data point within a distribution.
# 5. Comparisons: Z-scores enable comparisons between data points from different distributions, even if those distributions have different 
# means and standard deviations.
# 6. Hypothesis Testing: Z-scores are used in hypothesis testing to assess the significance of results and make informed decisions 
# based on sample data.

# Overall, z-scores provide a standardized framework for understanding and analyzing data points within the context of their respective distributions.
```

```python
# Question 9

# Answer 9 -
# The central limit theorem states that if you take sufficiently large samples from a population,
# the samples’ means will be normally distributed, even if the population isn’t normally distributed.
# According to the central limit theorem, the mean of a sample of data will be closer to the mean of the overall population in question, 
# as the sample size increases, notwithstanding the actual distribution of the data
```

```python
# Queation 10
# Answer 10 -

# The assumptions of the Central Limit Theorem in bullet format:

# 1. Independence: The random variables in the sample must be independent of each other.
# 2. Sample Size: The sample size should be sufficiently large. There is no strict rule for the minimum sample size, but a commonly cited 
# guideline is that the sample size should be at least 30.
# 3. Population Distribution: The population from which the samples are drawn does not have to be normal, but it should have a finite mean 
# and variance (or a large enough sample size).
# 4. Sampling Variability: The sample means should have similar variances regardless of the population distribution, 
# provided the sample size is large enough.
```
