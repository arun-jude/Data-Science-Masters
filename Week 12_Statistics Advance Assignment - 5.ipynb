```python
# Question 2

# Answer 2 -

import scipy.stats as st

expected_data=[20,20,20,10,10,20]
observed_data=[25,15,15,15,15,15]

## Chi square Goodness OF Fit
chisquare_test_statistics,p_value=st.chisquare(observed_data,expected_data)
print("Chi Square Statistic is: ",chisquare_test_statistics)
print("P-Value is: ",p_value)

## find the critical value
significance=0.05
dof=len(observed_data)-1
critical_value=st.chi2.ppf(1-significance,dof)
print("Critical value is: ",critical_value)

print("Null Hypothesis is: H0= The observed data meets the expectation")
print("Alternate Hypothesis is: H1= The observed data does not meet the expectation")

if chisquare_test_statistics > critical_value:
    print("Reject the null hypothesis (H0)")
else:
    print("Accept the null hypothesis (H0)")
```

```python
# Question 4

# Answer 4 -

import numpy as np
from scipy.stats import norm

# Given values
sample_size = 500
sample_proportion = 60 / sample_size  # Calculating the sample proportion

# Calculate the standard error
standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Find the critical value for a 95% confidence level
confidence_level = 0.95
z_critical = norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_critical * standard_error

# Calculate the confidence interval
confidence_interval_lower = sample_proportion - margin_of_error
confidence_interval_upper = sample_proportion + margin_of_error

print("Sample Proportion:", sample_proportion)
print("Margin of Error:", margin_of_error)
print("Confidence Interval:", (confidence_interval_lower, confidence_interval_upper))

```

```python
# Question 9

# Answer 9 -

# Null Hypothesis, H0= smoking status and lung cancer diagnosis are independent,
# Alternate Hypothesis, H1 = there is a significant association between smoking status and lung cancer diagnosis

# Observed (smoker,yes)=60
# Observed (smoker,no)=140
# Observed (nonsmoker,yes)=30
# Observed (nonsmoker,yes)=170

# Expected (smoker,yes)= (200 * 90)/400 = 45
# Expected (smoker,no)= (200*310)/400 = 155
# Expected (nonsmoker,yes)=(200 * 90)/400 = 45
# Expected (nonsmoker,yes)= (200*310)/400 = 155


# Chi-square (smoker,yes) = ((Observed - Expected)**2)/Expected = ((60-45)**2)/45 = 5
# Chi-square (smoker,no) = ((Observed - Expected)**2)/Expected = ((140-155)**2)/155 = 1.452
# Chi-square (nonsmoker,yes) = ((Observed - Expected)**2)/Expected = ((30-45)**2)/45 = 5
# Chi-square (nonsmoker,no) = ((Observed - Expected)**2)/Expected = ((170-155)**2)/155 = 1.452

# Overall Chi-square value = sum of all the above chi-square values = 5 + 1.452 + 5 + 1.452 = 12.904

# degree of freedom df = (2-1)*(2-1)=1

# alpha = 0.05

# basis obove degree of freedom & alpha value, chi square table value = 3.84

# calculated test statistic of 12.904 > chi square table value of 3.84. Hence, we reject the null hypothesis
# Final Conclusion : There is a significant association between smoking status and lung cancer diagnosis
```
